```python
import pandas as pd

# Function to parse a line of the HMMER output
def parse_hmmscan_line(line):
    parts = line.strip().split()
    if len(parts) < 22:
        return None  # Skip lines that don't have the minimum number of fields
    # Parse the parts into a dictionary
    return {
        "target_name": parts[0],
        "accession": parts[1],
        "tlen": int(parts[2]),
        "query_name": parts[3],
        "query_accession": parts[4],
        "qlen": int(parts[5]),
        "full_seq_E-value": float(parts[6]),
        "full_seq_score": float(parts[7]),
        "full_seq_bias": float(parts[8]),
        "domain_num": int(parts[9]),
        "total_domains": int(parts[10]),
        "domain_c-Evalue": float(parts[11]),
        "domain_i-Evalue": float(parts[12]),
        "domain_score": float(parts[13]),
        "domain_bias": float(parts[14]),
        "hmm_start": int(parts[15]),
        "hmm_end": int(parts[16]),
        "ali_start": int(parts[17]),
        "ali_end": int(parts[18]),
        "env_start": int(parts[19]),
        "env_end": int(parts[20]),
        "acc": float(parts[21]),
        "description": " ".join(parts[22:])
    }

# Read the HMMER output file
hmmscan_file = "results.domtblout"
parsed_lines = []

with open(hmmscan_file, 'r') as file:
    for line in file:
        if not line.startswith('#') and line.strip():
            parsed_line = parse_hmmscan_line(line)
            if parsed_line:
                parsed_lines.append(parsed_line)

# Convert parsed lines to a DataFrame
pfam_results = pd.DataFrame(parsed_lines)

# Display the DataFrame
pfam_results.head()
```
